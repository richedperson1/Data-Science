```python
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
## Q-Q plot


x = [i for i in range(100)]
y = [i*30 for i in range(100)]
```

```python
plt.plot(y,x)
```

```python
def calculateM_C(x,y):

    n = len(x)

    sum_xy = sum(xi * yi for xi, yi in zip(x, y))
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x_squared = sum(xi ** 2 for xi in x)

    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    c = (sum_y - m * sum_x) / n

    print("m =", m)
    print("c =", c)
    angle_rad = math.atan(m)

    # Step 3: Convert the angle to degrees if necessary
    angle_deg = math.degrees(angle_rad)

    print("Angle in degrees:", angle_deg)
```

```python

x = [i for i in range(100)]
y = [i*90 for i in range(100)]

calculateM_C(x,y)
plt.plot(y,x,label = "Multiple : 90")
y = [i*80 for i in range(100)]
plt.plot(y,x,label = "Multiple : 80")
y = [i*10 for i in range(100)]
plt.plot(y,x ,label = "Multiple : 10")

y = [i*2 for i in range(100)]
plt.plot(y,x,label = "Multiple : 2")
plt.legend()
plt.show()
```
