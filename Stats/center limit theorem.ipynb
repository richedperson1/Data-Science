## Center Limit theorem proof

```python
"""
CLT work with any kind of distribution no matter what it will be !
"""
```

```python
from scipy.stats import skewnorm
import matplotlib.pyplot as plt

numValues = 100000
maxValue = 100
skewness = -5   #Negative values are left skewed, positive values are right skewed.

random = skewnorm.rvs(a = skewness,loc=maxValue, size=numValues)  #Skewnorm function

random = random - min(random)      #Shift the set so the minimum value is equal to zero.
random = random / max(random)      #Standadize all the vlues between 0 and 1. 
random = random * maxValue         #Multiply the standardized values by the maximum value.

#Plot histogram to check skewness

sns.kdeplot(random)
```

```python
# Center limit theorem


```

```python
def centerLimitTheorem(sampleSize=None):
    sampleMean = np.array([])
    sampleStd  = np.array([])
    if not sampleSize:
        sampleSize = 30
    iterationCount = 10000
    for ind in range(iterationCount):
        sampleArr = np.random.choice(random,sampleSize)
        sampleMean = np.append(sampleMean,np.mean(sampleArr))
        sampleStd = np.append(sampleStd,np.std(sampleArr))


    print(f"Standard deviation is : {np.std(sampleMean)}, sample size {sampleSize}")
    sns.kdeplot(sampleMean,label = f"sample size {sampleSize}")
    plt.title("Sample mean Distribution")
    plt.legend()
    plt.show()
    return sampleMean
```

```python
num = 5
element = centerLimitTheorem(num)
print("Population standard devition : ",np.std(random),"Sample stadard devition * sqrt(n) : ",np.std(element)*np.sqrt(num))
```
