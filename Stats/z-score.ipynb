```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Generate random numbers from a log-normal distribution
mu = 5  # Mean of the underlying normal distribution
sigma = 3  # Standard deviation of the underlying normal distribution
size = 1000  # Number of samples
samples = np.random.lognormal(mu, sigma, size)

# Plot a histogram of the generated samples
sns.kdeplot(samples)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Random Log-Normal Distribution')
plt.grid(True)

# Display the plot
plt.show()
```

```python
z_scores = stats.zscore(samples)
sns.kdeplot(z_scores)
```
