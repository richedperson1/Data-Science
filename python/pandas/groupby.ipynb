### Library importing

```python
import pandas as pd
from datetime import datetime as dt
import numpy as np
```

### Data loading

```python
RawDataset = pd.read_csv("../../dataset/orders.csv")
RawDataset.head()
```

```python
RawDataset["datetime"] = pd.to_datetime(RawDataset.date)
RawDataset["month_ind"] = RawDataset.datetime.map(lambda x : x.month)
RawDataset.head()
```
